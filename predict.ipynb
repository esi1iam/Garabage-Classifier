```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
from tensorflow.keras.layers import BatchNormalization
```

```python
model = tf.keras.models.load_model('garabage_classifier.h5')
```

```python
from tensorflow.keras.preprocessing import image
import cv2
output_class = ["Battery", "Biological", "Brown Glass", "Cardboard", "Clothes", "Green Glass",
                "Metal", "Paper", "Plastic ", "Shoes", "Trash", "white-glass" ]

def waste_prediction(IMG):
  img = tf.keras.preprocessing.image.load_img(IMG, target_size=(300, 300))
  resize = tf.image.resize(img, (256,256))
  plt.imshow(resize.numpy().astype(int))
  plt.show()
  img_array = tf.keras.preprocessing.image.img_to_array(img)
  img_array = tf.expand_dims(img_array, 0) 
  
  predictions = model.predict(img_array)
  
  #print(predictions[0]*100, "\n", output_class )
  print("Prediction: ", output_class [np.argmax(predictions)], f"{predictions[0][np.argmax(predictions)]*100}%")
```
